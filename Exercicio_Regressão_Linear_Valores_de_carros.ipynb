```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
import seaborn as sns
```

#Leitura dos dados

```python
df = pd.read_csv('carro.csv')
df.head()
```

```python
df.describe()
```

#Dumificação dos dados

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df.head()
```

```python
#Dumificação dos dados de variáveis categóricas para numéricas

le.fit(df.drivewheel)
df.drivewheel = le.transform(df.drivewheel)

le.fit(df.fuelsystem)
df.fuelsystem = le.transform(df.fuelsystem)

le.fit(df.enginetype)
df.enginetype = le.transform(df.enginetype)

le.fit(df.enginelocation)
df.enginelocation = le.transform(df.enginelocation)

le.fit(df.CarName)
df.CarName = le.transform(df.CarName)

le.fit(df.fueltype)
df.fueltype = le.transform(df.fueltype)

le.fit(df.aspiration)
df.aspiration = le.transform(df.aspiration)

le.fit(df.carbody)
df.carbody = le.transform(df.carbody)
```

```python
df.describe()
```

```python
df.head()
```

#Alteração manual de valores de dados

A alteração manual se faz importante para identificar mais facilmente as informações

```python
#alteração manual da quantidade de portas para ficar mais entendível o dado:

df['doornumber'] = df['doornumber'].replace(['two'], '2')
df['doornumber'] = df['doornumber'].replace(['four'], '4')
```

```python
df1 = df.drop(columns= ['car_ID', 'symboling', 'CarName', 'fueltype', 'aspiration', 'doornumber'])
df1.head()
```

```python
df['cylindernumber'].value_counts()
```

```python
#alteração manual da quantidade de cilindros para melhor identificação da informação

df['cylindernumber'] = df['cylindernumber'].replace(['two'], '2')
df['cylindernumber'] = df['cylindernumber'].replace(['three'], '3')
df['cylindernumber'] = df['cylindernumber'].replace(['four'], '4')
df['cylindernumber'] = df['cylindernumber'].replace(['five'], '5')
df['cylindernumber'] = df['cylindernumber'].replace(['six'], '6')
df['cylindernumber'] = df['cylindernumber'].replace(['eight'], '8')
df['cylindernumber'] = df['cylindernumber'].replace(['twelve'], '12')
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

#Normalização do dataframe

```python
#normalização do dataframe
df_norm = pd.DataFrame(scaler.fit_transform(df), index = df.index, columns = df.columns)
df_norm.head()
```

#Criação das variáveis de teste e treino

Educando o modelo

```python
x = df_norm[['CarName', 'drivewheel', 'enginelocation', 'carwidth', 'carheight','cylindernumber','enginesize','boreratio','stroke','compressionratio','horsepower']]

y = df_norm[['price']]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)

lr = LinearRegression()

lr.fit(x,y)
```

```python
r_sqr = lr.score(x, y)
r_sqr
```

```python
y_pred_treino = lr.predict(x_train)
print('MAE:', metrics.mean_absolute_error(y_train, y_pred_treino))
print('MSE:', metrics.mean_squared_error(y_train, y_pred_treino))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_train, y_pred_treino)))
```

```python
y_pred_test = lr.predict(x_test)
print('MAE:', metrics.mean_absolute_error(y_test, y_pred_test))
print('MSE:', metrics.mean_squared_error(y_test, y_pred_test))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred_test)))
```
